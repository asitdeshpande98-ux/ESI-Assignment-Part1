After Importing our new dataset, we can compute the Absolute wind speed as discussed in the research proposal, y using pythagoras equiation.

```python
#add a new variable to dataset final: absolute wind speed based on the ERA5. 
#Adding it to the dataframe also provides us with the U100 and V100 if we still require them, and computing and new dataset is more time consuming
dataset_final['ws_100'] = np.sqrt(
    dataset_final['u100']**2 + dataset_final['v100']**2
)


#Using this absolute windspeed at 100m, we can use the equation provided to us by : I.lazarat et al. (2024)
#ratio of the height to the power of roughness of terrian
# v2/v1 = (h2/h1)**a     with a= terrain roughness (open sea is 0.1)

hub_height_turbine= 250
hub_height_standard= 100
terrain_roughness= 0.1

# scale wind speed from 100m to 250m
dataset_final['ws_250'] = dataset_final['ws_100'] * (hub_height_turbine / hub_height_standard) ** terrain_roughness
```

Now we also determine the wind direction which can be useful later and also shows us interctions with its envirment

```python
import numpy as np

# --- 1. Wind direction (degrees) ---
dataset_final['wind_dir'] = (
    180 + np.degrees(np.arctan2(dataset_final['u100'], dataset_final['v100']))
) % 360

# Optional: mask calm winds
dataset_final['wind_dir'] = dataset_final['wind_dir'].where(
    dataset_final['ws_100'] > 0
)

# --- 2. Convert to 8 sectors (FAST) ---
# Shift by 22.5 so bins center on N, NE, etc.
sector_index = ((dataset_final['wind_dir'] + 22.5) // 45) % 8

# Map index → direction label
labels = np.array(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])

dataset_final['wind_dir_cat'] = xr.DataArray(
    labels[sector_index.astype(int)],
    dims=dataset_final['wind_dir'].dims,
    coords=dataset_final['wind_dir'].coords
)

# --- 3. Quick check ---
print(dataset_final[['wind_dir', 'wind_dir_cat']])
print("Unique sectors:", np.unique(dataset_final['wind_dir_cat']))
```

Ny now finishing the base interpertation of the wind, additional factor that apply to the wind speed can applied upon the model.
In the next part the model will calculate the wake effect, which will act as a coefficeint upon the windspeed, resulting in a final useable windspeed.
By additional complexitity in the model this part may become much larger

```python

# wake loss coefficients (10% each)
intra_wake_loss = 0.10
inter_wake_loss = 0.10

# apply wake effects to wind speed at 250 m
dataset_final['ws_250_wake'] = (
    dataset_final['ws_250'] 
    * (1 - intra_wake_loss) 
    * (1 - inter_wake_loss)
)
```

With the final windspeed in place, the capacity factor based upon the turbines wind profile can be determined. 
Using data specific to a certain turbine, a specific capacity factor and Annual energy production be be calculated

```python
#formula to determine capacity factor of wind, based on the above given description. 
def calculate_wind_cf(data, cut_in=4, rated=13, cut_off=25, smooth=False):
    ws = data["ws_250_wake"]

    cf = xr.zeros_like(ws)

    if smooth:
        ramp_cf = ((ws - cut_in) / (rated - cut_in)) ** 2
    else:
        ramp_cf = (ws - cut_in) / (rated - cut_in)

    # Ramp region
    cf = xr.where(
        (ws >= cut_in) & (ws < rated),
        ramp_cf,
        cf
    )

    # Rated region
    cf = xr.where(
        (ws >= rated) & (ws <= cut_off),
        1.0,
        cf
    )

    # Above cut-off
    cf = xr.where(
        ws > cut_off,
        0.0,
        cf
    )

    return cf

dataset_final["cf_wind"] = calculate_wind_cf(
    dataset_final,
    cut_in=4,
    rated=13,
    cut_off=25,
    smooth=False
)

print(dataset_final["cf_wind"].min().values)
print(dataset_final["cf_wind"].max().values)

dataset_final[["ws_100", "ws_250", "ws_250_wake", "cf_wind"]]

dataset_final["ws_100"].plot()
plt.title("Wind speed at 100 m")
plt.show()

dataset_final["ws_250"].plot()
plt.title("Wind speed at 250 m")
plt.show()

dataset_final["ws_250_wake"].plot()
plt.title("Wind speed at 250 m after wake losses")
plt.show()

dataset_final["cf_wind"].plot()
plt.title("Wind Capacity Factor")
plt.show()

smooth=True

dataset_final["cf-wind"] = calculate_wind_cf(dataset_final, smooth=True)

#sanity check for results, 250m windspeed should be 9-10% higher
#based upon the a factor provided
#sanity check, the 250_wake should be smaller  then the 250m due to wake losses
```

As a result we now have the capacity factor for every point in the dataset, giving us annual energy production, when multiplied by the rated power for the specific turbine model on which the curve was already based.

```python
# two ways to get towards the final ANNUAL energy production.


# 1: simplest and seems best fit for out model so far. This is simple both by far more effective
rated_power = 15e6  # 15 MW turbine (example)

dataset_final['power_output'] = rated_power * dataset_final['cf_wind']


#2: formula, but also doing double work

air_density = 1.225  # kg/m^3
rotor_diameter = 200  # meters (example large offshore turbine)
radius = rotor_diameter / 2
area = np.pi * radius**2

cp = 0.45  # realistic efficiency (NOT full Betz)

dataset_final['power_raw'] = (
    0.5
    * air_density
    * area
    * cp
    * dataset_final['ws_250_wake']**3
)

```

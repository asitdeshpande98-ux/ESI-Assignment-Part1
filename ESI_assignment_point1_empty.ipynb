```python
# Function to import climate data from JRC - DO NOT CHANGE

def import_jrc_climate_data(lon: float, lat: float, year: int | str, alt: float) -> dict:
    """
    Reads in climate data for a full year from `JRC PVGIS <https://re.jrc.ec.europa.eu/pvg_tools/en/>`_.

    :param float lon: longitude of node - the api will read data for this location
    :param float lat: latitude of node - the api will read data for this location
    :param int year: optional, needs to be in range of data available. If nothing is specified, a typical year \
    will be loaded
    :param float alt: altitude of location specified
    :return: dict containing information on the location (altitude, longitude, latitude and a dataframe \
    containing climate data (ghi = global horizontal irradiance, dni = direct normal irradiance, \
    dhi = diffuse horizontal irradiance, rh = relative humidity, temp_air = air temperature, ws = wind speed at \
    specified hight. Wind speed is returned as a dict for different heights.
    :rtype: dict
    """

    # Specify year import, lon, lat
    parameters = {"lon": lon, "lat": lat, "year": year, "outputformat": "json"}
    time_index = pd.date_range(
        start=str(year) + "-01-01 00:00", end=str(year) + "-12-31 23:00", freq="1h"
    )

    # Get data from JRC dataset
    answer = dict()
    print("Importing Climate Data...")
    response = requests.get("https://re.jrc.ec.europa.eu/api/tmy?", params=parameters)
    if response.status_code == 200:
        print("Importing Climate Data successful")
    else:
        raise Exception(response)
    data = response.json()
    climate_data = data["outputs"]["tmy_hourly"]

    # Compile return dict
    answer["longitude"] = lon
    answer["latitude"] = lat
    answer["altitude"] = alt

    ghi = []
    dni = []
    dhi = []
    rh = []
    temp_air = []
    wind_speed = dict()
    wind_speed["10"] = []

    for t_interval in climate_data:
        ghi.append(t_interval["G(h)"])
        dni.append(t_interval["Gb(n)"])
        dhi.append(t_interval["Gd(h)"])
        rh.append(t_interval["RH"])
        temp_air.append(t_interval["T2m"])
        wind_speed["10"].append(t_interval["WS10m"])

    answer["dataframe"] = pd.DataFrame(
        np.array([ghi, dni, dhi, temp_air, rh]).T,
        columns=["ghi", "dni", "dhi", "temp_air", "rh"],
        index=time_index,
    )
    for ws in wind_speed:
        answer["dataframe"]["ws" + str(ws)] = wind_speed[ws]

    return answer
```

Financial Functions

```python
def annuity(r, n):
    if r == 0:
        return 1 / n
    return r / (1 - (1 + r) ** (-n))

def monthly_annualized_cost(capex, lifetime, fixed_opex_fraction=0):
    return capex * (annuity(discount_rate, lifetime) + fixed_opex_fraction) / 12

wind_monthly_cost_per_MW = monthly_annualized_cost(wind_capex, wind_lifetime, wind_opex) * 1000
solar_monthly_cost_per_MW = monthly_annualized_cost(solar_capex, solar_lifetime, solar_opex) * 1000
line_monthly_cost_per_MW = monthly_annualized_cost(line_capex * distance12, line_lifetime, line_opex) * 1000
battery_monthly_cost_per_MWh = monthly_annualized_cost(battery_capex, battery_lifetime, battery_opex)

line_loss_fraction = line_losses_per_100km * distance12 / 100
```

Prepare Time Series

```python
T = len(times)

demand = np.asarray(el_demand, dtype=float)                   # MW each hour
cf_wind = np.asarray(capacity_factor_wind, dtype=float)       # pu
cf_solar = np.asarray(capacity_factor_solar, dtype=float)     # pu

cf_wind = np.nan_to_num(cf_wind, nan=0.0)
cf_solar = np.nan_to_num(cf_solar, nan=0.0)
demand = np.nan_to_num(demand, nan=0.0)

# Candidate line sizes in 10 MW increments
max_line_candidate = int(np.ceil((demand.max() + 20) / 10.0) * 10)
line_candidates = np.arange(0, max_line_candidate + 10, 10)
```

Solver

```python

def solve_model(storage_location="zone2", line_cap=50):
    """
    storage_location = 'zone1' or 'zone2'
    line_cap = fixed electricity cable size in MW
    """

    # --------------------------------------------------------
    # Variable layout
    # --------------------------------------------------------
    # Scalar variables:
    # 0 = wind capacity [MW]
    # 1 = solar capacity [MW]
    # 2 = battery energy capacity [MWh]
    #
    # Hourly variables:
    # send[t]       : power sent from zone 1 into line [MW]
    # charge[t]     : battery charging power [MW]
    # discharge[t]  : battery discharging power [MW]
    # soc[t]        : battery state of charge [MWh]
    # import[t]     : external grid import in zone 2 [MW]
    # shed[t]       : unserved demand [MW]

    idx_wind = 0
    idx_solar = 1
    idx_batt = 2

    start_send = 3
    start_charge = start_send + T
    start_discharge = start_charge + T
    start_soc = start_discharge + T
    start_import = start_soc + T
    start_shed = start_import + T

    n_vars = start_shed + T

    def send_i(t): return start_send + t
    def charge_i(t): return start_charge + t
    def discharge_i(t): return start_discharge + t
    def soc_i(t): return start_soc + t
    def import_i(t): return start_import + t
    def shed_i(t): return start_shed + t

    # --------------------------------------------------------
    # Objective
    # --------------------------------------------------------
    c = np.zeros(n_vars)

    c[idx_wind] = wind_monthly_cost_per_MW
    c[idx_solar] = solar_monthly_cost_per_MW
    c[idx_batt] = battery_monthly_cost_per_MWh

    for t in range(T):
        c[import_i(t)] = electricity_import_cost
        c[shed_i(t)] = lost_load_penalty

    fixed_line_cost = line_cap * line_monthly_cost_per_MW

    # --------------------------------------------------------
    # Bounds
    # --------------------------------------------------------
    bounds = [(0, None)] * n_vars

    for t in range(T):
        bounds[send_i(t)] = (0, line_cap)
        bounds[charge_i(t)] = (0, None)
        bounds[discharge_i(t)] = (0, None)
        bounds[soc_i(t)] = (0, None)
        bounds[import_i(t)] = (0, max_import)
        bounds[shed_i(t)] = (0, None)

    # --------------------------------------------------------
    # Equality constraints
    # --------------------------------------------------------
    A_eq = []
    b_eq = []

    # 1) Demand balance at zone 2
    for t in range(T):
        row = np.zeros(n_vars)

        # Delivered power after line losses
        row[send_i(t)] = (1 - line_loss_fraction)
        row[import_i(t)] = 1.0
        row[shed_i(t)] = 1.0

        if storage_location == "zone2":
            row[discharge_i(t)] = 1.0
            row[charge_i(t)] = -1.0

        A_eq.append(row)
        b_eq.append(demand[t])

    # 2) Storage SOC balance
    for t in range(T):
        row = np.zeros(n_vars)

        row[soc_i(t)] = 1.0
        row[charge_i(t)] = -eta_charge
        row[discharge_i(t)] = 1.0 / eta_discharge

        if t == 0:
            # Start at 50% SOC
            row[idx_batt] = -0.5
        else:
            row[soc_i(t - 1)] = -(1 - self_discharge)

        A_eq.append(row)
        b_eq.append(0.0)

    A_eq = np.array(A_eq)
    b_eq = np.array(b_eq)

    # --------------------------------------------------------
    # Inequality constraints
    # --------------------------------------------------------
    A_ub = []
    b_ub = []

    # 1) Renewable availability at zone 1
    # send + charge(zone1) <= wind + solar
    for t in range(T):
        row = np.zeros(n_vars)
        row[send_i(t)] = 1.0
        row[idx_wind] = -cf_wind[t]
        row[idx_solar] = -cf_solar[t]

        if storage_location == "zone1":
            row[charge_i(t)] = 1.0

        A_ub.append(row)
        b_ub.append(0.0)

    # 2) If storage is in zone1, sent power can also include discharge from local battery
    # send <= wind + solar + discharge - charge
    # Rearranged as: send + charge - discharge - wind - solar <= 0
    if storage_location == "zone1":
        for t in range(T):
            row = np.zeros(n_vars)
            row[send_i(t)] = 1.0
            row[charge_i(t)] = 1.0
            row[discharge_i(t)] = -1.0
            row[idx_wind] = -cf_wind[t]
            row[idx_solar] = -cf_solar[t]

            A_ub.append(row)
            b_ub.append(0.0)

    # 3) Charge power limit
    for t in range(T):
        row = np.zeros(n_vars)
        row[charge_i(t)] = 1.0
        row[idx_batt] = -max_charge_frac
        A_ub.append(row)
        b_ub.append(0.0)

    # 4) Discharge power limit
    for t in range(T):
        row = np.zeros(n_vars)
        row[discharge_i(t)] = 1.0
        row[idx_batt] = -max_discharge_frac
        A_ub.append(row)
        b_ub.append(0.0)

    # 5) SOC upper bound
    for t in range(T):
        row = np.zeros(n_vars)
        row[soc_i(t)] = 1.0
        row[idx_batt] = -1.0
        A_ub.append(row)
        b_ub.append(0.0)

    A_ub = np.array(A_ub)
    b_ub = np.array(b_ub)

    # --------------------------------------------------------
    # Solve
    # --------------------------------------------------------
    res = linprog(
        c=c,
        A_ub=A_ub,
        b_ub=b_ub,
        A_eq=A_eq,
        b_eq=b_eq,
        bounds=bounds,
        method="highs"
    )

    if not res.success:
        return None

    x = res.x

    result = {
        "storage_location": storage_location,
        "line_cap_MW": line_cap,
        "total_cost_EUR_month": res.fun + fixed_line_cost,
        "wind_MW": x[idx_wind],
        "solar_MW": x[idx_solar],
        "battery_MWh": x[idx_batt],
        "send_MW": x[start_send:start_send + T],
        "charge_MW": x[start_charge:start_charge + T],
        "discharge_MW": x[start_discharge:start_discharge + T],
        "soc_MWh": x[start_soc:start_soc + T],
        "import_MW": x[start_import:start_import + T],
        "shed_MW": x[start_shed:start_shed + T],
        "fixed_line_cost_EUR_month": fixed_line_cost
    }

    return result
```

Run Both Locations and Line Sizes

```python
all_results = []

for storage_loc in ["zone1", "zone2"]:
    for line_cap in line_candidates:
        res = solve_model(storage_location=storage_loc, line_cap=line_cap)
        if res is not None:
            all_results.append(res)

summary = pd.DataFrame([
    {
        "storage_location": r["storage_location"],
        "line_cap_MW": r["line_cap_MW"],
        "wind_MW": r["wind_MW"],
        "solar_MW": r["solar_MW"],
        "battery_MWh": r["battery_MWh"],
        "imports_MWh": np.sum(r["import_MW"]),
        "lost_load_MWh": np.sum(r["shed_MW"]),
        "total_cost_EUR_month": r["total_cost_EUR_month"]
    }
    for r in all_results
]).sort_values("total_cost_EUR_month").reset_index(drop=True)

print(summary.head(10))
```

Best Solution

```python
best = all_results[np.argmin([r["total_cost_EUR_month"] for r in all_results])]

print("\nBEST SOLUTION")
print(f"Storage location : {best['storage_location']}")
print(f"Line capacity    : {best['line_cap_MW']:.2f} MW")
print(f"Wind capacity    : {best['wind_MW']:.2f} MW")
print(f"Solar capacity   : {best['solar_MW']:.2f} MW")
print(f"Battery size     : {best['battery_MWh']:.2f} MWh")
print(f"Monthly cost     : {best['total_cost_EUR_month']:.2f} EUR")
print(f"Imports          : {np.sum(best['import_MW']):.2f} MWh")
print(f"Lost load        : {np.sum(best['shed_MW']):.6f} MWh")
```

```python
best_ts = pd.DataFrame(index=times)
best_ts["demand_MW"] = demand
best_ts["wind_gen_MW"] = best["wind_MW"] * cf_wind
best_ts["solar_gen_MW"] = best["solar_MW"] * cf_solar
best_ts["renewable_gen_MW"] = best_ts["wind_gen_MW"] + best_ts["solar_gen_MW"]
best_ts["line_send_MW"] = best["send_MW"]
best_ts["line_delivered_MW"] = best["send_MW"] * (1 - line_loss_fraction)
best_ts["battery_charge_MW"] = best["charge_MW"]
best_ts["battery_discharge_MW"] = best["discharge_MW"]
best_ts["battery_soc_MWh"] = best["soc_MWh"]
best_ts["grid_import_MW"] = best["import_MW"]
best_ts["lost_load_MW"] = best["shed_MW"]

if best["storage_location"] == "zone1":
    best_ts["curtailment_MW"] = (
        best_ts["renewable_gen_MW"]
        + best_ts["battery_discharge_MW"]
        - best_ts["battery_charge_MW"]
        - best_ts["line_send_MW"]
    ).clip(lower=0)
else:
    best_ts["curtailment_MW"] = (
        best_ts["renewable_gen_MW"]
        - best_ts["line_send_MW"]
    ).clip(lower=0)

print(best_ts.head())
```
